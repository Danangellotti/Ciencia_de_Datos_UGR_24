### Ejercicio 3b usando en groupby()

```python
# leemos dataset
ej3 = pd.read_csv("ejercicios/clima-multi.txt", sep="\s+")

#obtenemos un "dataframe agrupado" done se ponen columnas de groupby() como indices
grouped = ej3.groupby(["NOMBRE"])

print("## Promedio de humedades")
grouped["HUM"].mean()
```

### Ejercicio 3b usando indexación

```python
# leemos dataset
ej3 = pd.read_csv("ejercicios/clima-multi.txt", sep="\s+")

# obtenemos serie boolean con valores True para los rows
# que columna "NOMBRE" == "BARILOCHE_AERO"
rows_interes = ej3["NOMBRE"] == "BARILOCHE_AERO"

# nos quedamos solo con los rows donde rows_interes es True,
# osea solo los registros BARILOCHE_AERO
ej3_bari = ej3.loc[rows_interes]

# ahora en ej3_bari tenemos solo los registros de Bariloche
# calculamos el promedio de humedad de este subconjunto
ej3_bari_mean = ej3_bari["HUM"].mean()

print("## Humedad promedio Bariloche")
print(ej3_bari_mean)

# lo mismo para catamarca pero en un sola linea
ej3_cata_mean = ej3.loc[ej3["NOMBRE"] == "CATAMARCA_AERO"]["HUM"].mean()

print("## Humedad promedio Catamarca")
print(ej3_cata_mean)


```
