# [Git](https://git-scm.com/)

Es un sistema de control de versiones distribuido que rastrea los cambios en cualquier conjunto de archivos, normalmente utilizado para coordinar el trabajo entre programadores de forma colaborativa durante el desarrollo de software. Sus objetivos incluyen velocidad, integridad de datos y soporte para flujos de trabajo distribuidos y no lineales (miles de ramas paralelas ejecutándose en diferentes sistemas).

Al igual que la mayoría de los sistemas de control de versiones distribuidos, y a diferencia de la mayoría de los sistemas cliente-servidor, cada directorio Git en cada ordenador es un repositorio completo con un historial completo y capacidades de seguimiento de versiones, independientemente del acceso a la red o de un servidor central. Git es software libre y de código abierto distribuido bajo la licencia GPL-2.0-only.

Git es un ejecutable del sistema operativo es por este motivo que dentro del colab debemos utilizar el signo de admiración (!).

## Como maneja git los cambios

Cada vez que confirmas un cambio, o guardas el estado de tu proyecto, internamente se toma una foto del aspecto de todos tus archivos en ese momento y se guarda una referencia a esa copia instantánea. Para ser eficiente, si los archivos no se han modificado, no se almacena el archivo de nuevo, sino un enlace al archivo anterior idéntico que ya tiene almacenado. Git maneja sus datos como una secuencia de copias instantáneas.

image.png

## Operaciones locales

Git trabaja de forma local sin necesidad de estar conectandose a servidores. Es decir, una vez instalado podemos comenzar el traceo del nuestro código sin necesidad de estar conectados a internet. Solo necesitamos acceso a un servidor en caso de querer obtener la última versión del código o de tener que subir nuestros cambios.

## Integredidad

Todo en Git es verificado mediante checksum antes de ser almacenado, y es de esa forma como lo vamos a identificar en el futuro. Esto significa que es imposible cambiar el contenido de cualquier archivo o
directorio sin que Git lo sepa. Esta funcionalidad está integrada en Git al más bajo nivel y es parte integral de su filosofía. No puedes perder información durante su transmisión o sufrir corrupción de archivos sin que Git sea capaz de detectarlo.

El mecanismo que usa Git para generar este checksum se conoce como
hash SHA-1. Se trata de una cadena de 40 caracteres hexadecimales (0-9 y a-f), y se calcula con base en los contenidos del archivo o estructura del directorio en Git. Un hash SHA-1 se ve de la siguiente forma `24b9da6552252987aa493b52f8696cd6d3b00373`.


## Los 3 estados

Los 3 estados principales en los que se pueden encontrar tus archivos son :
> `confirmado` (committed): significa que los
datos están almacenados de manera segura en tu base de datos local.

> `modificado` (modified): significa que has modificado el archivo pero todavía no lo has confirmado a tu base de datos.

> `preparado` (staged): significa que has marcado un archivo modificado en su versión actual para que vaya en tu próxima confirmación.

Esto nos lleva a las tres secciones principales de un proyecto de Git: El directorio de Git (Git directory), el directorio de trabajo (working directory), y el área de preparación (staging area).

El directorio de Git es donde se almacenan los metadatos y la base de datos de objetos para tu proyecto. Es la parte más importante de Git, y es lo que se copia cuando clonas un repositorio desde otra computadora.

El directorio de trabajo es una copia de una versión del proyecto. Estos archivos se sacan de la base de datos comprimida en el directorio de Git, y se colocan en disco para que los puedas usar o modificar.

El área de preparación es un archivo, generalmente contenido en tu directorio de Git, que almacena información acerca de lo que va a ir en tu próxima confirmación. A veces se le denomina índice (“index”), pero se está convirtiendo en estándar el referirse a ella como el área de preparación.

El flujo de trabajo básico es:
1. Modificas una serie de archivos en tu directorio de trabajo.
2. Preparas los archivos, añadiéndolos a tu área de preparación.
3. Confirmas los cambios, lo que toma los archivos tal y como están en el área de preparación y almacena esa copia instantánea de manera permanente en tu directorio de Git.

Si una versión concreta de un archivo está en el directorio de Git, se considera confirmada (committed). Si ha sufrido cambios desde que se obtuvo del repositorio, pero ha sido añadida al área de preparación, está preparada (staged). Y si ha sufrido cambios desde que se obtuvo del repositorio, pero no se ha preparado, está modificada (modified).

## ¿Qué significa HEAD en Git?

Un repositorio Git es una colección de objetos y referencias. Los objetos se relacionan entre sí, y las referencias apuntan a objetos y a otras referencias. Los objetos principales en un repositorio Git son los commits, pero otros objetos incluyen blobs y árboles. Las referencias más importantes en Git son las ramas, en las que puedes pensar como etiquetas que pones en un commit.

HEAD es otro tipo importante de referencia. El propósito de HEAD es hacer un seguimiento del punto actual en un repositorio Git. En otras palabras, HEAD responde a la pregunta: "¿Dónde estoy ahora mismo?".

Por ejemplo, cuando usas el comando log, ¿cómo sabe Git a partir de qué commit debe empezar a mostrar los resultados? HEAD proporciona la respuesta. Cuando creas un nuevo commit, su padre es indicado por el lugar al que HEAD apunta actualmente.

## HEAD detached

[Más información](https://www.cloudbees.com/blog/git-detached-head)

Acabas de ver que HEAD en Git es sólo el nombre de una referencia que indica el punto actual en un repositorio. Entonces, ¿qué significa que esté unido o separado?

La mayoría de las veces, HEAD apunta a un nombre de rama. Cuando añades un nuevo commit, la referencia de tu rama se actualiza para apuntar a ella, pero HEAD permanece igual. Cuando cambias de rama, HEAD se actualiza para apuntar a la rama a la que has cambiado. Todo esto significa que, en estos escenarios, HEAD es sinónimo de "el último commit en la rama actual".

## Ignorando archivo

Para esto debemos crear un archivo con el nombre `.gitignore`.

El contenido de este puede tener el nombre completo del archivo, de un directorio o un patrón que se debe obviar.

Las reglas sobre los patrones son:
- Ignorar las líneas en blanco y aquellas que comiencen con #.
- Emplear patrones glob estándar que se aplicarán recursivamente a todo el directorio del repositorio local.
- Los patrones pueden comenzar en barra (/) para evitar recursividad.
- Los patrones pueden terminar en barra (/) para especificar un directorio.
- Los patrones pueden negarse si se añade al principio el signo de exclamación (!).

Ejemplo:
```txt
# ignora los archivos terminados en .a
*.a

# pero no lib.a, aun cuando había ignorado los archivos terminados en .a en la línea anterior
!lib.a

# ignora unicamente el archivo TODO de la raiz, no subdir/TODO
/TODO

# ignora todos los archivos del directorio build/
build/

# ignora doc/notes.txt, pero no este: doc/server/arch.txt
doc/*.txt

# ignora todos los archivos .txt del directorio doc/
doc/**/*.txt
```

Vamos a ver un ejemplo de como ignorar el archivo comentarios.txt.

````python
# creamos el archivo comentarios.txt
!echo "Archivo para comentarios personales">>comentarios.txt
# estado del proyecto
!git status
# tenemos el archivo como no traceado
````

````python
# listamos el directorio
!ls -a
````

````python
# creamos el archivo .gitignore
!echo "# solo ignoramos el archivo comentarios.txt" >>.gitignore
!echo "comentarios.txt" >>.gitignore
# estado del proyecto
!git status
# ahora tenemos 1 archivo no traceado (.gitignore)
# el archivo comentarios.txt ya no aparece más porque está siendo ignorado
````

## Recuperando archivos viejos

````python
# con el siguiente comando vamos a ver cuales son los commit
!git log --follow -- LICENSE.md
# como podemos ver la última vez que se modificó el archivo fue en
# commit donde fue agregado
````

````python
# vamos a guardar el checksum del commit donde se eliminó el archivo
# estamos saltando 1 checksum (--skip 1) y estamos obteniendo la última
# modificación del archivo y lo guardamos en el archivo /tmp/deleted_commit
!git rev-list HEAD --skip 1 -n 1 -- LICENSE.md > /tmp/deleted_commit
````

````python
# vamos restaurar el archivo desde el checksum que lo eliminó
!echo "La sentencia ejecutada es:"
# los archivos contienen los checksum para que se pueda replicar en distintas
# ejecuciones
!echo "!git checkout $(cat /tmp/deleted_commit) -- LICENSE.md"
!echo ""
!echo ""
!git checkout $(cat /tmp/deleted_commit) -- LICENSE.md
# estado del proyecto
!git status
````

````python
# listamos el directorio
!ls -a
````

````python
# vamos a agregar el archivo nuevamente
!git add LICENSE.md
# hacemos el commit
!git commit -m "Restauración del archivo de licencia."
# estado del proyecto
!git status
# aparece una cartel raro, HEAD detached
````
