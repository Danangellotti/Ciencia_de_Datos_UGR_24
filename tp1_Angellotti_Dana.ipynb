b) Cargar el archivo en un DataFrame y mostrar el contenido

```python
import pandas as pd

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv("datos_clima.csv")

# Mostrar el contenido del DataFrame
df.head()  # Ver las primeras filas del archivo
```

c) Mostrar las columnas con su tipo de datos

```python
# Muestra las columnas y sus tipos de datos
df.info()
```

d) Eliminar las columnas que tienen todos valores nulos

```python
# Elimina las columnas donde todos los valores son nulos
df = df.dropna(axis=1, how='all')

# Mostrar las columnas restantes
df.info()
```

e) Identificar y eliminar filas duplicadas

```python
# Identifica si hay filas duplicadas
print(f"Filas duplicadas antes de eliminar: {df.duplicated().sum()}")

# Elimina filas duplicadas
df = df.drop_duplicates()

# Verifica de nuevo
print(f"Filas duplicadas después de eliminar: {df.duplicated().sum()}")
```

f) Identificar si hay valores nulos restantes

```python
# Identifica si hay valores nulos
print("Valores nulos por columna:\n", df.isnull().sum())
```

g) Reemplazar los valores nulos por 0

```python
# Reemplaza valores nulos por 0
df.fillna(0, inplace=True)

# Verifica si quedan valores nulos
print("Valores nulos después del reemplazo:\n", df.isnull().sum())
```

h) Generar algunas medidas descriptivas

```python
# Medida descriptivas del DataFrame
df.describe()
```

```python
# Filtra solo las columnas numéricas
df_numerico = df.select_dtypes(include=['number'])

# Calcula el promedio de las columnas numéricas
df_numerico.mean()
```
