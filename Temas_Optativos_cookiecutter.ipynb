```
!cookiecutter -c v1 https://github.com/drivendata/cookiecutter-data-science -f --output-dir /content
```

Comando y argumentos por partes:

* __\-c v1 https://github.com/drivendata/cookiecutter-data-science__: contiene la vesión _v1_ y el repositorio de donde se va a descargar la estructura del proyecto.
* __\-f__: fuerza la creación de los directorios, se usa principalmente si el directorio ya existe.
* __\-\-output-dir /content__: directorio padre donde se creará la estructura.

Una vez que ejecutemos ese comando nos va a realizar las siguientes preguntas:


> __[1/8] project_name (project_name):__ Se debe ingresar el nombre del proyecto. Se va a crear un directorio con este nombre dentro del directorio que se pasa como parámetro al comando cookiecutter (\-\-output-dir).

> __[2/8] repo_name (proyecto_data_science):__ nombre del repositorio que utilizaremos

> __[3/8] author_name (Your name (or your organization/company/team)):__ nombre de la empresa o nombre del equipo o nombre del desarrollador que lleva a cabo el desarrollo.

> __[4/8] description (A short description of the project.):__ breve descripción del proyecto.

> __[5/8] Select open_source_license\
    1 - MIT\
    2 - BSD-3-Clause\
    3 - No license file\
    Choose from [1/2/3] (1):__ 1\
    Debemos seleccionar un tipo de licencia del proyecto las 2 primeras son de software libre, la número 3 es la licencia propietaria. \
    Por defecto viene seleccionada la número 1 o MIT.\
    [Licencia MIT](https://www.mit.edu/~amini/LICENSE.md)\
    [Licencia BSD-3-clause](https://stat.ethz.ch/CRAN/web/licenses/BSD_3_clause)

> __[6/8] s3_bucket ([OPTIONAL] your-bucket-for-syncing-data (do not include 's3://')):__ configuración de un bucket de aws, por defecto dejar en blanco.

> __[7/8] aws_profile (default):__ configuración de un bucket de aws, por defecto dejar en blanco.

> __[8/8] Select python_interpreter\
    1 - python3\
    2 - python\
    Choose from [1/2] (1):__ 1\
    Seleccionar el interprete de Python que se va a utilizar.\
    Por defecto viene seleccionado el 1 o Python3

## Archivos generados

### Makefile

Si desean ejecutar o actualizar una tarea cuando se actualizan determinados archivos, la utilidad make puede venir muy bien. La utilidad make requiere de un archivo, makefile, que define un conjunto de tareas a ejecutar.

A muchos de ustedes les sonara este nombre pues puede que hayan utilizado make para compilar un programa desde código fuente. La mayoría de los proyectos de código abierto utilizan make para compilar un binario ejecutable final, que puede ser instalado usando make install.

En esta ocasión vamos a entender cómo es que funciona este tipo de archivos dentro de Linux.

Un archivo makefile básicamente se distingue en cuatro tipos básicos de declaraciones:

* Comentarios
* Variables.
* Reglas explicitas.
* Reglas implícitas.

Las Reglas explicitas le indican a make que archivos dependen de otros archivos, así como los comandos requeridos para compilar un archivo en particular.

Mientras que las implícitas son similares que las explicitas, pero con la diferencia indican los comandos a ejecutar, sino que make utiliza las extensiones de los archivos para determinar que comandos ejecutar.

### README.md

Lenguaje de marcado ligero que se utiliza para escribir archivos de documentación.

Para más información visitar [markdownguide](https://www.markdownguide.org/).

Actualmente se está popularizando [asciidoc](https://asciidoctor.org/) porque tiene más detalles del marcado, estos archivos poseen una extensión `adoc`.

### requirements.txt

Archivos que contiene todas las librerías o módulos que debemos instalar para poder ejecutar el programa o proyecto.

### setup.py

Este archivo me permite hacer que un programa o proyecto sea instalable en el sistema operativo y de esta forma lo pueda ejecutar como cualquier otro programa.

Pueden ampliar con [setuptools](https://setuptools.pypa.io/en/latest/userguide/index.html)

### tox.ini

Archivo de configuración para la automatización de tareas.

Pueden ampliar con [tox](https://tox.wiki/en/4.14.2/)
