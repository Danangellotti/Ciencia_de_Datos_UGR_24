## Solución Ejemplo 1: Distribución de Frecuencias



```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos del ejemplo
datos = ['E', 'D', 'D', 'E', 'C', 'C', 'D', 'E', 'C', 'D', 'D', 'D', 'D', 'C', 'C', 'D', 'E', 'D', 'C', 'D',
         'D', 'E', 'E', 'C', 'D', 'E', 'E', 'D', 'C', 'D', 'C', 'E', 'D', 'C', 'C', 'D', 'E', 'D', 'C', 'C',
         'D', 'D', 'C', 'D', 'C', 'D', 'C', 'D', 'D', 'D']

# Crear DataFrame
df = pd.DataFrame({'Forma_de_Pago': datos})

# Calcular frecuencias
freq_abs = df['Forma_de_Pago'].value_counts().sort_index()
freq_rel = freq_abs / len(datos)
freq_abs_acum = freq_abs.cumsum()
freq_rel_acum = freq_rel.cumsum()

# Crear tabla de frecuencias
tabla_freq = pd.DataFrame({
    'Frecuencia Absoluta': freq_abs,
    'Frecuencia Relativa': freq_rel,
    'Frecuencia Absoluta Acumulada': freq_abs_acum,
    'Frecuencia Relativa Acumulada': freq_rel_acum
})

# Mostrar tabla
print(tabla_freq)

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
freq_abs.plot(kind='bar')
plt.title('Distribución de Formas de Pago')
plt.xlabel('Forma de Pago')
plt.ylabel('Frecuencia Absoluta')
plt.show()
```

Este código crea una tabla de distribución de frecuencias y un gráfico de barras para el Ejemplo 1. Utiliza pandas para el manejo de datos y matplotlib para la visualización.

## Solución Ejemplo 2: Cantidad de Hijos de Empleados

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos del ejemplo
datos = [0, 1, 1, 4, 2, 2, 1, 3, 2, 3, 1, 2, 3, 0, 2]

# Crear DataFrame
df = pd.DataFrame({'Cantidad_Hijos': datos})

# Calcular frecuencias
freq_abs = df['Cantidad_Hijos'].value_counts().sort_index()
freq_rel = freq_abs / len(datos)
freq_abs_acum = freq_abs.cumsum()
freq_rel_acum = freq_rel.cumsum()

# Crear tabla de frecuencias
tabla_freq = pd.DataFrame({
    'Frecuencia Absoluta': freq_abs,
    'Frecuencia Relativa': freq_rel,
    'Frecuencia Absoluta Acumulada': freq_abs_acum,
    'Frecuencia Relativa Acumulada': freq_rel_acum
})

# Mostrar tabla
print(tabla_freq)

# Crear gráfico de bastones
plt.figure(figsize=(10, 6))
plt.stem(freq_abs.index, freq_abs.values,use_line_collection=True)
plt.title('Distribución de Cantidad de Hijos')
plt.xlabel('Cantidad de Hijos')
plt.ylabel('Frecuencia Absoluta')
plt.xticks(freq_abs.index)
plt.show()

# Crear gráfico escalonado para frecuencias acumuladas
plt.figure(figsize=(10, 6))
plt.step(freq_abs_acum.index, freq_abs_acum.values, where='post')
plt.title('Frecuencia Absoluta Acumulada')
plt.xlabel('Cantidad de Hijos')
plt.ylabel('Frecuencia Absoluta Acumulada')
plt.xticks(freq_abs_acum.index)
plt.show()
```

## Solución Ejemplo 3: Edades Agrupadas en Intervalos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos del ejemplo
datos = [62, 54, 68, 46, 38, 54, 58, 64, 76, 56, 42, 48, 52, 72, 66, 56, 32, 56, 62, 58]

# Calcular el número de clases y el ancho de cada clase
n = len(datos)
num_clases = int(np.sqrt(n))
ancho_clase = (max(datos) - min(datos)) / num_clases

# Crear los límites de los intervalos (bin edges)
# Instead of using pd.interval_range, we generate bin edges using np.linspace
bin_edges = np.linspace(min(datos), max(datos), num_clases + 1)

# Agrupar los datos en intervalos
df = pd.DataFrame({'Edad': datos})
df['Intervalo'] = pd.cut(df['Edad'], bins=bin_edges)

# Calcular frecuencias
freq_abs = df['Intervalo'].value_counts().sort_index()
freq_rel = freq_abs / n
freq_abs_acum = freq_abs.cumsum()
freq_rel_acum = freq_rel.cumsum()

# Crear tabla de frecuencias
tabla_freq = pd.DataFrame({
    'Frecuencia Absoluta': freq_abs,
    'Frecuencia Relativa': freq_rel,
    'Frecuencia Absoluta Acumulada': freq_abs_acum,
    'Frecuencia Relativa Acumulada': freq_rel_acum
})

# Mostrar tabla
print(tabla_freq)

# Crear histograma
# Pass bin_edges instead of intervalos to plt.hist
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=bin_edges, edgecolor='black')
plt.title('Histograma de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia Absoluta')
plt.show()

# Crear polígono de frecuencias
plt.figure(figsize=(10, 6))
freq_abs.plot(kind='line', marker='o')
plt.title('Polígono de Frecuencias')
plt.xlabel('Edad')
plt.ylabel('Frecuencia Absoluta')
plt.xticks(rotation=45)
plt.show()
```
