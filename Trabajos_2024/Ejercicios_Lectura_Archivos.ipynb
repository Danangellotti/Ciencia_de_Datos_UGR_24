### Ejercicio 3

El archivo `ejercicios/clima-multi.txt` es similar al clima.txt, salvo que tiene informaci√≥n de dos estaciones de metereolog√≠a. La tarea es informar el promedio de humedad (HUM) de los dos lugares.

* a) Explorar el archivo con comandos del shell
* b) Generar un dataframe con el archivo, almacenarlo en la variable ej3 y mostrar su contenido
* c) Calcular y mostrar el promedio de Humedad de las dos estaciones.


_Ayuda_: Si se crea el dataframe usando `index_col`, se pueden 'segmentar' los datos accediendolos con `loc[...]`

```python
# 3a)
# Muestra todo el contenido del archivo
!cat ejercicios/clima-multi.txt

# tu respuesta
```

```python
# 3b)
# Cargar el archivo clima-multi.txt, usando la primera columna como √≠ndice (index_col=0)
ej3 = pd.read_csv('ejercicios/clima-multi.txt', delimiter='\s+', index_col=0)

# Mostrar el contenido del DataFrame
ej3.head()
# tu respuesta
```

```python
# 3c)
# Calcular el promedio de la columna 'HUM' (Humedad)
promedio_humedad = ej3['HUM'].mean()

# Mostrar el promedio de humedad de las dos estaciones
print(f"El promedio de la humedad de las dos estaciones es: {promedio_humedad}")
# tu respuesta
```
