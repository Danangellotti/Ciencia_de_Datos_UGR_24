# Predicción del precio de casas usando regresión lineal : 📉



* import numpy as np: Importa la librería NumPy, que se utiliza para trabajar con arreglos (arrays) y realizar operaciones matemáticas eficientes.
* import matplotlib.pyplot as plt: Importa la librería Matplotlib, que se utiliza para crear gráficos en Python.

* from mpl_toolkits.mplot3d import Axes3D: Importa Axes3D de Matplotlib, que permite crear gráficos tridimensionales.





```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```



*   Aquí se crean tres arrays de NumPy, x1, x2, y x3, que representan las características de diferentes casas:

      *   x1: Tamaño de la casa en metros cuadrados.
      *   x2: Número de habitaciones en la casa.
      *   x3: Ubicación de la casa, categorizada en tres valores diferentes.



```python
# Ejemplo de datos
x1 = np.array([100, 150, 200, 250, 300])  # Tamaño (en metros cuadrados)
x2 = np.array([2, 3, 4, 4, 5])            # Número de habitaciones
x3 = np.array([1, 2, 3, 2, 1])            # Ubicación (categorizada como 1, 2, 3)
```



*   Se definen valores arbitrarios para los pesos (w1, w2, w3) y el bias (b) en el modelo de regresión lineal. Estos valores se usarán para calcular la predicción del precio.

  *   w1: Peso asignado a la característica de tamaño.
  *   w2: Peso asignado a la característica del número de habitaciones.
  *   w3: Peso asignado a la característica de la ubicación.
  *   b: Bias, una constante que se suma al final de la predicción.





```python
# Pesos y bias arbitrarios para la regresión lineal
w1, w2, w3 = 500, 10000, 20000
b = 50000
```



*   Aquí se calcula el precio predicho (y) para cada casa utilizando la fórmula de regresión lineal:
  *   y = w1*x1 + w2*x2 + w3*x3 + b: Se multiplica cada característica por su peso correspondiente y se suma el bias para obtener el precio predicho.



```python
# Predicción del precio
y = w1*x1 + w2*x2 + w3*x3 + b
```



*   Se crea una figura (fig) con un tamaño especificado de 10x8 pulgadas.
*   ax = fig.add_subplot(111, projection='3d'): Se añade un subplot a la figura con una proyección 3D, lo que permite dibujar gráficos tridimensionales.



```python
# Gráfico 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
```



*   ax.scatter: Se dibujan los puntos en el gráfico 3D utilizando las características x1, x2, y x3 como coordenadas en el espacio tridimensional.

  *   c=y: El color de los puntos se asigna en función del valor de y (el precio predicho), utilizando el mapa de colores viridis.
  *   marker='o': Define que los puntos se dibujen como círculos.

 *   s=100: Define el tamaño de los puntos.



```python
# Dibujar los puntos
ax.scatter(x1, x2, x3, c=y, cmap='viridis', marker='o', s=100)
```



*   Se etiquetan los ejes del gráfico 3D
  *   ax.set_xlabel('Tamaño (m²)'): Etiqueta el eje X como "Tamaño (m²)".
  *   ax.set_ylabel('Número de habitaciones'): Etiqueta el eje Y como "Número de habitaciones".
  *   ax.set_zlabel('Ubicación (categoría)'): Etiqueta el eje Z como "Ubicación (categoría)".



```python
# Etiquetas de los ejes
ax.set_xlabel('Tamaño (m²)')
ax.set_ylabel('Número de habitaciones')
ax.set_zlabel('Ubicación (categoría)')
```



* Se añade un título al gráfico: "Predicción del precio de casas usando regresión lineal".




```python
# Título del gráfico
ax.set_title('Predicción del precio de casas usando regresión lineal')
```



* plt.show(): Muestra el gráfico en pantalla.

  Este código crea un gráfico tridimensional que visualiza cómo un modelo de regresión lineal predice los precios de las casas en función de sus características (tamaño, número de habitaciones y ubicación).



```python
plt.show()
```
