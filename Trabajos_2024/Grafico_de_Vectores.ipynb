### Descripción del código
Importaciones: Importamos numpy para crear los vectores y matplotlib.pyplot para graficar.
Definición de vectores: Creamos dos vectores vector_a y vector_b.
Gráfico: Usamos plt.quiver para graficar los vectores desde el origen (0, 0) hacia las coordenadas definidas por los vectores.
Configuración del gráfico: Establecemos límites, etiquetas y una cuadrícula para mejorar la visualización.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los vectores
vector_a = np.array([2, 3])  # Vector A
vector_b = np.array([4, 1])  # Vector B

# Crear un gráfico
plt.figure()
plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector A (2, 3)')
plt.quiver(0, 0, vector_b[0], vector_b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector B (4, 1)')

# Configurar límites y etiquetas
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.title('Gráfico de Vectores')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
