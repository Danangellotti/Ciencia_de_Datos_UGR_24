# Predicci칩n del precio de casas usando regresi칩n lineal : 游늴



* import numpy as np: Importa la librer칤a NumPy, que se utiliza para trabajar con arreglos (arrays) y realizar operaciones matem치ticas eficientes.
* import matplotlib.pyplot as plt: Importa la librer칤a Matplotlib, que se utiliza para crear gr치ficos en Python.

* from mpl_toolkits.mplot3d import Axes3D: Importa Axes3D de Matplotlib, que permite crear gr치ficos tridimensionales.





```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```



*   Aqu칤 se crean tres arrays de NumPy, x1, x2, y x3, que representan las caracter칤sticas de diferentes casas:

      *   x1: Tama침o de la casa en metros cuadrados.
      *   x2: N칰mero de habitaciones en la casa.
      *   x3: Ubicaci칩n de la casa, categorizada en tres valores diferentes.



```python
# Ejemplo de datos
x1 = np.array([100, 150, 200, 250, 300])  # Tama침o (en metros cuadrados)
x2 = np.array([2, 3, 4, 4, 5])            # N칰mero de habitaciones
x3 = np.array([1, 2, 3, 2, 1])            # Ubicaci칩n (categorizada como 1, 2, 3)
```



*   Se definen valores arbitrarios para los pesos (w1, w2, w3) y el bias (b) en el modelo de regresi칩n lineal. Estos valores se usar치n para calcular la predicci칩n del precio.

  *   w1: Peso asignado a la caracter칤stica de tama침o.
  *   w2: Peso asignado a la caracter칤stica del n칰mero de habitaciones.
  *   w3: Peso asignado a la caracter칤stica de la ubicaci칩n.
  *   b: Bias, una constante que se suma al final de la predicci칩n.





```python
# Pesos y bias arbitrarios para la regresi칩n lineal
w1, w2, w3 = 500, 10000, 20000
b = 50000
```



*   Aqu칤 se calcula el precio predicho (y) para cada casa utilizando la f칩rmula de regresi칩n lineal:
  *   y = w1*x1 + w2*x2 + w3*x3 + b: Se multiplica cada caracter칤stica por su peso correspondiente y se suma el bias para obtener el precio predicho.



```python
# Predicci칩n del precio
y = w1*x1 + w2*x2 + w3*x3 + b
```



*   Se crea una figura (fig) con un tama침o especificado de 10x8 pulgadas.
*   ax = fig.add_subplot(111, projection='3d'): Se a침ade un subplot a la figura con una proyecci칩n 3D, lo que permite dibujar gr치ficos tridimensionales.



```python
# Gr치fico 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
```



*   ax.scatter: Se dibujan los puntos en el gr치fico 3D utilizando las caracter칤sticas x1, x2, y x3 como coordenadas en el espacio tridimensional.

  *   c=y: El color de los puntos se asigna en funci칩n del valor de y (el precio predicho), utilizando el mapa de colores viridis.
  *   marker='o': Define que los puntos se dibujen como c칤rculos.

 *   s=100: Define el tama침o de los puntos.



```python
# Dibujar los puntos
ax.scatter(x1, x2, x3, c=y, cmap='viridis', marker='o', s=100)
```



*   Se etiquetan los ejes del gr치fico 3D
  *   ax.set_xlabel('Tama침o (m)'): Etiqueta el eje X como "Tama침o (m)".
  *   ax.set_ylabel('N칰mero de habitaciones'): Etiqueta el eje Y como "N칰mero de habitaciones".
  *   ax.set_zlabel('Ubicaci칩n (categor칤a)'): Etiqueta el eje Z como "Ubicaci칩n (categor칤a)".



```python
# Etiquetas de los ejes
ax.set_xlabel('Tama침o (m)')
ax.set_ylabel('N칰mero de habitaciones')
ax.set_zlabel('Ubicaci칩n (categor칤a)')
```



* Se a침ade un t칤tulo al gr치fico: "Predicci칩n del precio de casas usando regresi칩n lineal".




```python
# T칤tulo del gr치fico
ax.set_title('Predicci칩n del precio de casas usando regresi칩n lineal')
```



* plt.show(): Muestra el gr치fico en pantalla.

  Este c칩digo crea un gr치fico tridimensional que visualiza c칩mo un modelo de regresi칩n lineal predice los precios de las casas en funci칩n de sus caracter칤sticas (tama침o, n칰mero de habitaciones y ubicaci칩n).
