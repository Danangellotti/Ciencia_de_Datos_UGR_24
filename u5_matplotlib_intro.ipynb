### Dos interfaces


#### Interfaz estilo MATLAB

Basada en estado. Mantiene un 'seguimiento' del estado actual de la figure y sus ejes, el estado va cambiando ante cada uso de los comandos plt.
Se peude recuperar el estado acutal con `plt.gcf` (figura actual) y `plt.gca` (ejes acutales).

````python
# un array de 100 numeros de 0 a 10
x = np.linspace(0, 10, 100)

plt.figure()  # create a plot figure
plt.suptitle("Una Figura")

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));

````

#### Interfaz Orientada a objetos
No se basa en el concepto de una figura o eje activo.

Se basa en poder invocar a métodos que cambian el estado de cualquier figura o ejes.

````python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));

# podemos cambiar el primer grafico usando ax[0]
fig.suptitle("Una Figura")
ax[0].set_title("Titulo 1")
ax[1].set_title("Titulo 2")
````

## Ajustar colores de lineas y estilos

````python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # hex code (RRGGBB, 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # HTML color names supported
````

## Matplotlib: plt vs axes   

Por lo general plt y axes tienen los mismos metodos. Pero hay ciertos atributps que no.

- `plt.xlabel`  → `ax.set_xlabel`
- `plt.ylabel` → `ax.set_ylabel`
- `plt.xlim`  → `ax.set_xlim`
- `plt.ylim` → `ax.set_ylim`
- `plt.title` → `ax.set_title`

Para cambiar estos atributos `Axes` tienen el método `set`

````python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10),
       ylim=(-2, 2),
       xlabel='x',
       ylabel='sin(x)',
       title='A Simple Plot');
````

# Scatter Plots: plt.scatter

En los gráficos `scatter` (dispersión, representar la relacion de dos variables) solo se dibujan los puntos representados por puntos, circulos y u otras figuras.

`plt.scatter` es otra forma más específica para este tipo de gráficos:

````python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.scatter(x, y, marker='o');
````

Con `plt.scatter` se pueden controlar las propiedades de cada punto (size, face color, edge color, etc.)

````python
rng = np.random.default_rng(0)
x = rng.normal(size=100)
y = rng.normal(size=100)
colors = np.random.randint(0, 100, 100)
sizes = 1000 * rng.random(100)

# agregamos el grafico scatter
plt.scatter(x, y, c=colors, s=sizes, alpha=0.4)

# agregamos la barra de colores de referencia
plt.colorbar();
````

El argumeto de `color` es mapeado automaticamente a una escala de colores (que se puede ver llamando a `colorbar`). También ver que el argumento `size` es dado en pixels.

## Usar alpha para solapar y comparar distribuciones

````python
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)
````

````python
plt.hist(x1, histtype='bar', alpha=0.5, bins=20)  #25
plt.hist(x2, histtype='bar', alpha=0.5, bins=20)  #26
plt.hist(x3, histtype='bar', alpha=0.5, bins=20);  #24
````

````python
plt.hist(x1, histtype='step', alpha=0.3, bins=40, linewidth=2)  #25
plt.hist(x2, histtype='step', alpha=0.3, bins=40, linewidth=2)  #26
plt.hist(x3, histtype='step', alpha=0.3, bins=40, linewidth=2);  #24
````

Cual es el mejor numero de `bins`. Usando un string se pueden usar distintos estimadores para calcular el valor óptimo.  [hitogram bin edges](https://numpy.org/doc/stable/reference/generated/numpy.histogram_bin_edges.html#numpy.histogram_bin_edges)

````python
plt.hist(x1, histtype='stepfilled', alpha=0.3, bins='auto')
plt.hist(x2, histtype='stepfilled', alpha=0.3, bins='auto')
plt.hist(x3, histtype='stepfilled', alpha=0.3, bins='auto');
````

## Histogramas de dos dimensiones

Asi como creamos histogramas en una dimension. podemos crear histogramas en dos dimensiones diviendo las cantidades de puntos en bins en los dos ejes.

Usamos un hist2d para mostrar la distribuciones de dos variables cuando el número de valores es múy grande.

````python
rng = np.random.default_rng(1701) # generador con seed 1701

x = rng.normal(0, 1, 1000)
y = rng.normal(0, 1, 1000)
````

### plt.hist2d: Histograma de dos dimensiones

````python
#plt.style.use("default")

plt.hist2d(x, y, bins=15)
cb = plt.colorbar()
cb.set_label('Conteo de contenedores (bins)')
````

numpy también un metodo para calcular un histograma 2d `np.histogram2d`:

````python
counts, xedges, yedges = np.histogram2d(x, y, bins=15)

print("Conteos")
print(counts)

print("\nBordes")
print(xedges)
print(yedges)

````

### plt.hexbin: Contenedores Hexagonales

hist2d muestra los datos en una mosaico de cuadrados.
otra manera de completar un plano es usar mosaicos hexagonales. Para esto se puede usar `plt.hexbin`

````python
plt.hexbin(x, y, gridsize=15)
cb = plt.colorbar(label='conteo')
````

# Lineas verticales, horizontales y de referencia

Con `axvline` y `axhline` podemos trazar lineas horizontales y verticales.

````python
# ejemplo de linea vertical
plt.axvline(1.5, color="g", linestyle="--", linewidth=2);
````

````python
# Ejemplo de lina horizontal
plt.axhline(1.5, color="b", linestyle="--", linewidth=2);
````

Ejemplo de como usarla para indicar referencias. Son utiliz para indicar márgenes, corrimientos, máximos o mínimos.

````python
m = -0.4
x1 = rng.normal(m, 1, 1000)

# dibujamos histograma
h = plt.hist(x1, histtype='stepfilled', alpha=0.3, density=True, bins='auto')

# calculamos media de las muestras y mostramos en verde
x1_mean = np.mean(x1)
plt.axvline(x1_mean, color="g", linestyle="--", linewidth=1);


# marcamos el cero como punto de referencia
plt.axvline(0, color="#444", linestyle=":", linewidth=2);


# calculamos densidad maxima y mostramos en azul
dmax = np.max(h[0])
plt.axhline(dmax, color="b", linestyle="--", linewidth=1);

````

````python
# dibujamos histograma
h = plt.hist(x1, histtype='stepfilled', alpha=0.3, density=True, bins='auto')

# calculamos media de las muestras y mostramos en verde
x1_mean = np.mean(x1)
plt.axvline(x1_mean, color="g", linestyle="--", linewidth=1);


# calculamos densidad maxima y mostramos en azul
dmax = np.max(h[0])
plt.axhline(dmax, color="b", linestyle="--", linewidth=1);

# agregamos anotaciones
print("mean:", x1_mean)

plt.annotate(f'mean: {x1_mean:.{3}f}',
            xy=(x1_mean, 0),
            xytext=(-2.5, 0.05),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate(f'dMax: {dmax:.{3}f}',
            xy=(-0.75, dmax),
            xytext=(-2.5, dmax - dmax/8), # para ubicarlo en las y usamos una posicion relativa a dmax
            arrowprops=dict(facecolor='black', shrink=0.05))
````

# Ticks

Las figuras (Figure) tienen ejes (Axes). Cada axes tiene un par de objetos para controlar las marcas de los ejes de los bordes.

Cada eje (Axes) tiene los atributos `xaxis` and `yaxis`, que tienen todas las propiedades de lineas: 'ticks' y 'labels' que hace a los ejes cartesianos (Axis).

## Ticks Majores y Menores


````python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np

%matplotlib inline
````

````python
# ejemplo para ver ticks majores y menores
ax = plt.axes(xscale='log', yscale='log')
ax.set(xlim=(1, 1E3), ylim=(1, 1E3))
ax.grid(True);
````

El eje muestra ticks majores, ticks menores, y labels.

Los ticks y labels se controlan con objetos llamados `locator` y `formatter` que tienen cada `xaxis` e `yaxis`

````python
# ejemplo donde solo se ven los ticks majores
ax = plt.axes()
ax.set(xlim=(1, 1E3), ylim=(1, 1E3))
ax.grid(True); # hacemos visible las lineas los ticks
````

````python
print(ax.xaxis.get_major_locator())
print(ax.xaxis.get_minor_locator())
````

````python
print(ax.xaxis.get_major_formatter())
print(ax.xaxis.get_minor_formatter())
````

En este caso las locaciones se controlan con un objeto `AutoLocator` y los ticks menores estan en con `NullFormatter`. Algo similar ocurren con los formatters.
