# Ejercicio 02

Al dataset anterior agregarle una nueva columna llamada `nltk_sin_stopword` quitando los stop words de la columna `news`según nltk.


```python
import nltk

nltk.download('stopwords')
nltk_stopwords = nltk.corpus.stopwords.words('spanish')
```

```python
df["nltk_sin_stopword"] = df["news"].apply(
    lambda value: " ".join([word for word in nltk.tokenize.ToktokTokenizer().tokenize(value) if word not in nltk_stopwords]))
df
```

# Ejercicio 03

Al dataset anterior agregarle una nueva columna llamada `nltk_stemn_sin_stopword` cambiando las palabras de la columna `nltk_sin_stopword` por su stem según nltk.


```python
from nltk.stem import SnowballStemmer
snowballStemmer = SnowballStemmer('spanish')
```

```python
df["nltk_stemn_sin_stopword"] = df["nltk_sin_stopword"].apply(
    lambda value: " ".join([snowballStemmer.stem(word) for word in nltk.tokenize.ToktokTokenizer().tokenize(value)]))
df
```

# Ejercicio 04

Al dataset anterior agregarle una nueva columna llamada `spacy_sin_stopword` quitando los stop words de la columna `news` según spacy.


```python
# descargo el modelo
!python -m spacy download es_core_news_sm

# importo el modelo
import spacy
nlp_es_sm = spacy.load("es_core_news_sm")
```

```python
df["spacy_sin_stopword"] = df["news"].apply(
    lambda value: " ".join([str(word) for word in nlp_es_sm(value) if str(word).lower() not in nlp_es_sm.Defaults.stop_words]))
df
```

# Ejercicio 05

Al dataset anterior agregarle una nueva columna llamada `spacy_lemma_sin_stopword` cambiando las palabras de la columna `spacy_sin_stopword` por su lemma según spacy.


```python
df["spacy_lemma_sin_stopword"] = df["spacy_sin_stopword"].apply(
    lambda value: " ".join([word.lemma_ for word in nlp_es_sm(value)]))
df
```
