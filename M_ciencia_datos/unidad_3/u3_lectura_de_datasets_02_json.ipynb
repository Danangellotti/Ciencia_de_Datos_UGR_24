# JSON

Tipicamente los datos __json__ los podemos encontrar:

*  en cadenas de texto,
* archivos o pueden
* ser provistos por servicios de internet.

Cuando se procesan en `python` por lo general se almancenan, como paso previo, en un `str` y luego se parsean con la libreria `json` que trasforma una cadena en un objeto de python que nos va a permitir acceder a los datos de forma estructurada.

Alamcenamos en la variable `person` una cadeana que contiene los mismos datos de la persona pero en formato __json__.

````python
personstr = """
{"name": "Wes",
 "age": 38,
 "cities_lived": ["Akron", "Nashville", "New York", "San Francisco"],
 "pet": null,
 "siblings": [{"name": "Scott", "age": 34, "hobbies": ["guitars", "soccer"]},
              {"name": "Katie", "age": 42, "hobbies": ["diving", "art"]}]
}
"""
personstr
````

````python
type(personstr)
````

````python
print(personstr)
````

**El formato json puede representar:**

* Un valor: numerico o string 1, 1.3 o "Hola"

* Un Objeto: { "nombre": "Manuel", ..... }

* Una lista de números: [1, 2, 3, 4, 5]

* Una lista de strings: ["A", "B", "C", "D"]

* Una lista de objetos:
[
  {"nombre": "Juan",  "edad": 33},
  {"nombre": "Maria", "edad": 41}
]

* Una lista de una mezcla de numeros, strings y objetos


Importamos la libreria `json`, parseamos la cadena, y guardamos el resultado en `result`

````python
import json

result = json.loads(personstr)
result
````

En este caso `result` es de tipo `dict`

````python
type(result)
````

````python
# podemos consultar por sus keys
result["name"]
````

````python
# podemos consultar por sus keys
result["siblings"]
````

La función `dumps` hace lo inverso. Combierte un objeto de `python` en una string con formato __json__.

````python
asjson = json.dumps(result)
print(asjson)
````

````python
# dumps acepta indent

asjson = json.dumps(result, indent=4)
print(asjson)
````

`asjson` es tipo str

````python
type(asjson)
````

## Web APIs

Los sitios web exponen servivios accesibles via HTTPS en formato json. Veremos un caso de como consultar las incidencias (tickets) de un proyecto de github.

Repositorio de pandas en github: https://github.com/pandas-dev/pandas

Los issues pueden verse aquí: https://github.com/pandas-dev/pandas/issues

Los issues en formato json: https://api.github.com/repos/pandas-dev/pandas/issues

Estos mismos issues puede leerse por http en formato json.

La libreria `requests` facilita realizar peticiones HTTP/S

Elegimos nuestra URL y realizamos la petición

````python
import requests

url = "https://api.github.com/repos/pandas-dev/pandas/issues"
````

````python
#pd.read_json(url)
````

````python
resp = requests.get(url)
````

Un valor de `status_code` de `200` significa petición exitosa

````python
resp.status_code
````

Si el request no fue exitoso, lanzar una excepción.

````python
resp.raise_for_status()
````

Podemos ver texto que envió el servicio

````python
## que hay en el cuerpo
print(resp.text)
````

````python
# para ver la cabecera
print(resp.headers)
````

Vemos el contenido del header en formato amistoso

````python
import json
print(json.dumps(dict(resp.headers), indent = 4))
````

Podemos parsear el body directamente a json con el método `json` del objeto `requests.response` que nos retorna un diccinario de python

````python
# exploramos los issues
issues_list = resp.json()
````

````python
## cuantos issues recuperamos
print(len(issues_list))
````

# XML

Tomamos como ejemplo la misma tabla:

     a   b   c
     1   2   3
     4   5   6
     7   8   9
    12  15  18

Estos datos se pueden representar en el siguiente `XML`

````python
simple_xml = """<?xml version="1.0" encoding="ISO-8859-1"?>
<datos>
    <row>
        <a>1</a>
        <b>2</b>
        <c>3</c>
    </row>
    <row>
        <a>4</a>
        <b>5</b>
        <c>6</c>
    </row>
    <row>
        <a>7</a>
        <b>8</b>
        <c>9</c>
    </row>

    <row>
        <a>12</a>
        <b>15</b>
        <c>18</c>
    </row>
</datos>
"""

print(simple_xml)
````

El formato `XML` puede leerse con [`read_xml`](https://pandas.pydata.org/docs/dev/reference/api/pandas.read_xml.html)

````python
simple = pd.read_xml(simple_xml)

simple
````

Los mismos datos pueden representarse segun este `XML`, haciendo uso de los `atributos` de los elementos.

````python
attrs_xml = """<?xml version="1.0" encoding="ISO-8859-1"?>
<data>
    <row a="1" b="2" c="3" />
    <row a="4" b="5" c="6" />
    <row a="7" b="8.0" c="xx" />
    <row a="12" b="15" c="18" />
</data>
"""

print(attrs_xml)
````

````python
attrs = pd.read_xml(attrs_xml)
attrs
````

````python
print("--a--")
print(attrs["a"], "\n")

print("--b--")
print(attrs["b"], "\n")

print("--c--")
print(attrs["c"], "\n")
````

# HTML

## Formato HTML

Tomamos como ejemplo la siguiente tabla

     a   b   c
     1   2   3
     4   5   6
     7   8   9
    12  15  18

Esa informacion se puede representar en formato `HTML` para que se muestre por un borwser asi:


<table>
            <thead>
                <tr>
                    <td>a</td><td>b</td><td>c</td>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>1</td><td>2</td><td>3</td>
                </tr>
                <tr>
                    <td>4</td><td>5</td><td>6</td>
                </tr>
                <tr>
                    <td>7</td><td>8</td><td>9</td>
                </tr>
            </tbody>
            <tfoot>
                <tr>
                    <td>12</td><td>15</td><td>18</td>
                </tr>
            </tfoot>
</table>

El formato `HTML` para representar esos datos, podría ser así:

``` html
<!DOCTYPE html>

<html>
<head>
<title>Title of the document</title>
</head>

<body>
    <h4>Hola html</h4>
    <table>
        <thead>
            <tr>
                <td>a</td><td>b</td><td>c</td>
            </tr>
        </thead>
        <tbody>
            <tr>
                <td>1</td><td>2</td><td>3</td>
            </tr>
            <tr>
                <td>4</td><td>5</td><td>6</td>
            </tr>
            <tr>
                <td>7</td><td>8</td><td>9</td>
            </tr>
        </tbody>
        <tfoot>
            <tr>
                <td>12</td><td>15</td><td>18</td>
            </tr>
        </tfoot>
     </table>
 </body>
```         

Por lo general la información en formato `HTML` de la web es más compleja que el ejemplo de arriba. Por ejemplo el archivo `examples/fdic_failed_bank_list.html` es una copia en achivos de [failed banklist](https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/)
