Para archivos muy grandes se puede ir procesando el archivo en **chunks** (simil a páginas) usando el parametro `chunksize`

```python
chunker = pd.read_csv("examples/ex6.csv", chunksize=1000)
type(chunker)
```

La siguiente celda calcula la cantidad de apariciones de cada valor de la columna **key** sin cargar el dataframe entero en memoria. El calculo se realiza cargando varios `piece` que se obtienen iterando sobre `chunker`. En este caso cada `piece` es un dataframe de 1000 rows.

```python
# obtenemos el dataframe en chunks de 1000 rows
chunker = pd.read_csv("examples/ex6.csv", chunksize=1000)

# creamos un serie para almacenar los resultados de conteo parciales
tot = pd.Series([], dtype='int64')

# iteramos sobre cada chunk. Y se va renovado el dataframe de a 1000 rows
for piece in chunker:
    # contamos los valores de key de cada chunk. Si existen NaN se interpretan como 0
    tot = tot.add(piece["key"].value_counts(), fill_value=0)

# mostramos el conteo de valores de 'key'
tot = tot.sort_values(ascending=False)
```

## Como cargar un dataframe en porciones e ir filtrando al mismo tiempo

```python
data = pd.read_csv("examples/ex6.csv")
data # <--- data tienen 10000rows y 5 columnas
```

Supongamos que solo necesitamos los key **"L" o "Q"**, y solo necesitamos las columnas **"one", "two" y "key"**.

Lo que podemos hacer es procesar todo el archivo cargando 'chunks' (bloques)
de 500 rows. Seleccionados los datos de interes del chunk y formamos un pequeño
dataframe temporal que tiene los datos de interes de ese chunk. A esta seleccion temporal dentro del bucle la agregamos al resultado final con `pd.concat`.

```python
# prepara el resultado
result = pd.DataFrame()

# procesamos todo el dataset cargando solo de a 500 rows
# por iteracion (vuelta del bucle)
iterator = pd.read_csv("examples/ex6.csv", chunksize=500)
for chunk in iterator:
    # que rows me quedo?
    # me interesan los rows que tengan key == L o Q
    # calculo una serie boolean de los row del chunk que cumplen esa condicion
    rows_selection = (chunk.key == "L") | (chunk.key == "Q")

    # que columnas me quedo?
    columns_selection = ["one", "two", "key"]

    # uso .loc[rows, cols]
    selection = chunk.loc[rows_selection, columns_selection]

    # voy agregando a result los selection que ya estan filtrados
    # por rows y por columns (mas pequeños que el chunk original)
    result = pd.concat([result, selection])


print("tamaño filtrado por chunks: ", len(result))

print(result) # <- el dataframe de interes es más
              #    pequeño( 686 rows y 3 columnas)
```
