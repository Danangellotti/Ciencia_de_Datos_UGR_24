## Combinar usando Merge

```python
df3
```

```python
tipos
```

```python
# left y right indican que dataframes combinar
# left_on y right_on indican sobre que columna hacer la combinacion
pd.merge(left=df3, right=tipos, left_on="tipo", right_on="key")
```

pd.merge: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html

```python
## usando suffixes podemos controlar el nombre de las columnas con nombres iguales
pd.merge(left=df3, right=tipos, left_on="tipo", right_on="key", suffixes=["", "_tipo"])
```

```python
# otra forma es ir renombrando las columnas
dfx = pd.merge(left=df3, right=tipos, left_on="tipo", right_on="key")
dfx
```

# Ejemplo Combinar usando merge

Ejemplo: Como unir dos dataframes de clientes y datos de personas para agregar más información referida a un cliente.

```python
df4
```

```python
!cat personas.csv
```

```python
personas = pd.read_csv("personas.csv")
personas
```

```python
# como ambos dataframe comparten el mismo nombre de columna podemos usar 'on='
pd.merge(left=df4, right=personas, on="dni")
```

Opciones de merge: inner, left, right, outer

```python
# how == "inner" (o inner join terminologia de las base de datos relacionales)
# preserva solo las filas que estan en ambos (es el default)
pd.merge(left=df4, right=personas, on="dni", how="inner")
```

```python
## left join (preserva las filas de left)
pd.merge(left=df4, right=personas, on="dni", how="left")
```

```python
## right (preserva las filas de right)
pd.merge(left=df4, right=personas, on="dni", how="right") ##, how="right")
```

```python
## outer join (preserva las filas de ambos)
dfo = pd.merge(left=df4, right=personas, on="dni", how="outer") ##, how="outer")
dfo
```

```python
## left join
dfo = pd.merge(left=df4, right=personas, on="dni", how="left")
dfo
```

Usamos left join, ya que queremos preservar la informacion de tipo y nombre de tipo

```python
# Nos quedamos con un left join
dfo = pd.merge(left=df4, right=personas, on="dni", how="left", suffixes=("_x", ""))
dfo
```
