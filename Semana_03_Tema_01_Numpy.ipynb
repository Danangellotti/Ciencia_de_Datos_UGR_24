## Acceso a los datos

Vamos a ver las distintas formas que tengo para acceder a los datos de un numpy array.

### Por posición

Para poder acceder se le pueden pasar valores fijos de fila y columna.

````python
print("Matriz base de trabajo (matriz): \n", matriz)
print()
print("Sentencia: matriz[0]")
print("Accedo a la fila 0 y obtengo el numpy array:", matriz[0], sep="\n")
print()
print("Sentencia: matriz[0, 2]")
print("Accedo a la fila 0 y columna 2 y obtengo el valor:",
      matriz[0, 2], sep="\n")
print()
print("Sentencia: matriz[0][2]")
print("Accedo a la fila 0 y columna 2 y obtengo el valor:",
      matriz[0][2], sep="\n")
print()
print("Sentencia: matriz.flatten[8]")
print("La función 'flatten' me devuelve un arreglo que une las filas.")
print("Accedo a un valor que esta en la posición 8 pero el acceso se realiza \n"
      "sobre un arreglo.", matriz.flatten()[8], sep="\n")
print()
print("Sentencia: matriz.flatten()")
print("Muestra del flat de la matriz: ", matriz.flatten())
````

### Por slicing

Siempre se hace desde el índice inferior incluido hasta el índice superior sin incluir.

Se pueden usar valores negativos donde se comienza desde el final hacia adelante.

El slicing puede tener 3 parámetros `indice_inferior:indice_superior:paso_o_incremento`.
- indice_inferior: es inclusivo.
- indice_superior: no se incluye.
- paso_o_incremento: por defecto es 1.

````python
print("Matriz base de trabajo (matriz):", matriz, sep="\n")
print()
print("Sentencia: matriz[1:3]")
print("Obtengo todas las filas que son mayores o iguales a 1 y menores que 3.\n"\
      "Se muestran las filas 1 y 2.",  matriz[1:3], sep="\n")
print()
print("Sentencia: matriz[:,2]")
print("Obtengo todas las filas de la columna 2.", matriz[:,2], sep="\n")
print()
print("Sentencia: matriz[:,1:3]")
print("Obtengo todas las filas y todas las columnas que son mayores o iguales \n"\
      "a 1 y menores que 3. Se muestran las columnas 1 y 2.",
      matriz[:,1:3], sep="\n")
print()
print("Sentencia: matriz[::2]")
print("Obtengo todas las filas con un incremento de 2. Se muestran las filas\n"\
      "pares o las filas 0,2,4 y todas las columnas.", matriz[::2], sep="\n")
print()
print("Sentencia: matriz[::2,::2]")
print("Obtengo todas las filas con un incremento de 2 y todas las columnas\n"\
      "con un incremento de 2. Se muestran las filas pares o las filas 0,2,4 y\n"\
      "las columnas pares o las columnas 0,2,4.", matriz[::2,::2], sep="\n")
````

### Por índices

Se accede a los valores de la matriz por arreglos de indices. En lugar de pasar una posición fija se pasan un arreglo de posiciones.

````python
# obtengo la cantidad de filas de la matriz
print("Cantidad de filas de la matriz:", matriz.shape[0], sep="\n")
print("Cantidad de columnas de la matriz:", matriz.shape[1], sep="\n")
print()
indices_filas = np.arange(start=0, stop=matriz.shape[0])
indices_columnas = np.arange(start=0, stop=matriz.shape[1])
indices_filas_invertidos = indices_filas[::-1]
indices_columnas_invertidos = indices_columnas[::-1]
print("Sentencia: np.arange(start=0, stop=matriz.shape[0])")
print("Creo el arreglo indices para las filas:", indices_filas, sep="\n")
print("Sentencia: np.arange(start=0, stop=matriz.shape[1])")
print("Creo el arreglo indices para las columnas:", indices_columnas, sep="\n")
print("Sentencia: indices_filas[::-1]")
print("Creo el arreglo indices para las filas invertido:",
      indices_filas_invertidos, sep="\n")
print("Sentencia: indices_columnas[::-1]")
print("Creo el arreglo indices para las columnas invertido:",
      indices_columnas_invertidos, sep="\n")
print()
print("Matriz base de trabajo (matriz):", matriz, sep="\n")
print()
print("Sentencia: matriz[indices_filas, indices_columnas]")
print(f"Muestro la diagonal que va desde la posición [0,0] hasta la posición \n"\
      f"[{matriz.shape[0]},{matriz.shape[1]}]:",
      matriz[indices_filas, indices_columnas], sep="\n")
print()
print("Sentencia: matriz[indices_filas, indices_columnas_invertidos]")
print(f"Muestro la diagonal que va desde la posición [0,{matriz.shape[1]}] \n"\
      f"hasta la posición :[{matriz.shape[0]},0]:",
      matriz[indices_filas, indices_columnas_invertidos], sep="\n")
print()
print("Sentencia: matriz[indices_filas_invertidos, indices_columnas_invertidos]")
print(f"Muestro la diagonal que va desde la posición [{matriz.shape[0]},{matriz.shape[1]}] \n"\
      f"hasta la posición :[0,0]:",
      matriz[indices_filas_invertidos, indices_columnas_invertidos], sep="\n")
print()
print("Sentencia: matriz[indices_filas_invertidos, indices_columnas]")
print(f"Muestro la diagonal que va desde la posición [{matriz.shape[0]},0] \n"\
      f"hasta la posición :[0,{matriz.shape[1]}]:",
      matriz[indices_filas_invertidos, indices_columnas], sep="\n")
````

## Broadcasting

Es la multidifusión de los objetos para poder operar con matrices. Esto me permite realizar operaciones entre una matriz NxM y

- Un entero
- Un arreglo
- Una matriz 1xM
- Una matriz Nx1

Python se encarga de generar nuevas matrices para poder operar de forma correcta.

`El broadcasting se realiza siempre!`


````python
# creo una matriz de 3 filas y 3 columnas
matriz = np.arange(start=1, stop=10).reshape((3, 3))
print("Matriz de trabajo (matriz):", matriz, sep="\n")

print()
print("Sentencia: np.add(matriz, 2)")
print("Suma de una matriz con un entero.",
      np.add(matriz, 2), sep="\n")
print("El valor 2 se transforma en la siguiente matriz ",
      [[2, 2, 2], [2, 2, 2], [2, 2, 2]], sep="\n")

print()
print("Sentencia: np.add(matriz, [10, 20, 30])")
print("Suma de una matriz y un vector.",
      np.add(matriz, [10, 20, 30]), sep="\n")
print("El vector [10, 20, 30] se transforma en la siguiente matriz ",
      [[10, 20, 30], [10, 20, 30], [10, 20, 30]], sep="\n")
print("IMPORTANTE: El arreglo debe tener la misma cantidad de columnas que la matriz.")

print()
print("Sentencia: np.add(matriz, [[100, 200, 300]])")
print("Suma de una matriz con otra matriz que posee solo 1 fila.",
      np.add(matriz, [[100, 200, 300]]), sep="\n")
print("Esta matriz [[100, 200, 300]] se transforma en la siguiente matriz",
      [[100, 200, 300], [100, 200, 300], [100, 200, 300]], sep="\n")
print("IMPORTANTE: Ambas matrices deben tener la misma cantidad de columnas")

print()
print("Sentencia: np.add(matriz, [[1000], [2000], [3000]])")
print("Suma de una matriz con otra matriz que solo posee 1 valor en cada fila.",
      np.add(matriz, [[1000], [2000], [3000]]), sep="\n")
print("Esta matriz [[1000], [2000], [3000]] se transforma en la siguiente matriz",
      [[1000, 1000, 1000],[2000, 2000, 2000],[3000, 3000, 3000]], sep="\n")
print("IMPORTANTE: Ambas matrices deben tener la misma cantidad de filas.")
````

## Unión de arreglos y matrices

Para unir arreglo vamos a utilizar la función `concatenate` que recibe como parámetro un objeto iterable de numpy array a concatenar y la dirección en que lo deseo hacer. Para la dirección se utiliza el argumento `axis`.

````python
# vamos a generar arreglos y matrices que utilizaremos más adelante
arreglo = np.arange(start=0, stop=12)
arreglo_1 = np.arange(start=1, stop=4)
arreglo_2 = np.arange(start=4, stop=7)

matriz_1 = np.arange(start=1, stop=7).reshape((2,3))
matriz_2 = np.arange(start=11, stop=17).reshape((2,3))
matriz_grande = np.concatenate([matriz_1, matriz_2])
matriz_grande = np.concatenate([matriz_grande, np.multiply(matriz_grande, 2)[::-1]])
matriz_grande = np.concatenate([matriz_grande, matriz_grande[::-1]], axis=1)
````

### Unión de arreglos


````python
print("Arreglo de trabajo (arreglo_1): \n", arreglo_1)
print("Arreglo de trabajo (arreglo_2): \n", arreglo_2)
print()
print("Sentencia: np.concatenate((arreglo_1, arreglo_2))")
print("Arreglos concatenados.",
      np.concatenate((arreglo_1, arreglo_2)), sep="\n")
````

### Unión de matrices

Tener cuidado con las dimensiones de las matrices:
- Si se agregan filas entonces deben poseer la misma cantidad de columnas.
- Si se agregan columnas entonces deben poseer la misma cantidad de filas.

````python
print("Matriz de trabajo (matriz_1):", matriz_1, sep="\n")
print("Matriz de trabajo (matriz_2):", matriz_2, sep="\n")
print()
print("Sentencia: np.concatenate((matriz_1, matriz_2))")
print("Matrices concatenadas agregando filas.",
      np.concatenate((matriz_1, matriz_2)), sep="\n")
print()
print("Sentencia: np.concatenate((matriz_1, matriz_2), axis=1)")
print("Matrices concatenadas agregando columnas.",
      np.concatenate((matriz_1, matriz_2), axis=1), sep="\n")
````

### Unión de arreglos y matrices

Para unir arreglos y matrices vamos a utilizar las siguientes funciones:
- `vstack`: poniendo el arreglo como una nueva fila.
- `hstack`: poniendo el arreglo como una nueva columna.


Tener cuidado con las dimensiones de las matrices:
- Si se agregan filas entonces deben poseer la misma cantidad de columnas.
- Si se agregan columnas entonces deben poseer la misma cantidad de filas.

````python
print("Arreglo de trabajo (arreglo_1):", arreglo_1, sep="\n")
print("Matriz de trabajo (matriz_2):", matriz_2, sep="\n")
print()
print("Sentencia: np.vstack((arreglo_1, matriz_2))")
print("Concatenación de matriz y arreglo por fila.",
      np.vstack((arreglo_1, matriz_2)), sep="\n")
print()
print("Necesito cambiar la forma del arreglo 1 para poder concatenarlos.")
arreglo_tmp = arreglo_1[0:2].reshape(2,1)
print("La nueva forma es [2,1]:", arreglo_tmp, sep="\n")
print()
print("Sentencia: np.hstack((arreglo_tmp, matriz_2))")
print("Concatenación de matriz y arreglo por columna.",
      np.hstack((arreglo_tmp, matriz_2)), sep="\n")
````

## Separación de arreglos y matrices


### Separación de un arreglo por índices

Para realizar la separación se utiliza la función `split` y el corte lo realiza para índices. Esta función recibe como parámetro un arreglo y una objeto iterable con los valores de los índices para realizar las separaciones.

````python
arreglo_tmp = np.arange(start=0, stop=12)
x_1, x_2, x_3, x_4 = np.split(arreglo_tmp, (3, 5, 8))

print("Arreglo de trabajo (arreglo_tmp):", arreglo_tmp, sep="\n")
print()
print("Sentencia: np.split(arreglo_tmp, (3, 5, 8))")
print()
print("x_1 (desde el índice 0 hasta el índice 3):", x_1, sep="\n")
print()
print("x_2 (desde el índice 3 hasta el índice 5):", x_2, sep="\n")
print()
print("x_3 (desde el índice 5 hasta el índice 8):", x_3, sep="\n")
print()
print("x_4 (desde el índice 8 hasta el final):", x_4, sep="\n")
````

### Separación de una matriz por filas

Para realizar la separación se utiliza la función `vsplit` y el corte lo realiza para índices. Esta función recibe como parámetro un matriz y una objeto iterable con los valores de los índices para realizar las separaciones.


````python
x_1, x_2, x_3, x_4 = np.vsplit(matriz_grande, (1, 3, 5))

print("Matriz de trabajo (matriz_grande):", matriz_grande, sep="\n")
print()
print("Sentencia: np.vsplit(matriz_tmp, (1, 3, 5))")
print()
print("x_1 (desde el índice 0 hasta el índice 1):", x_1, sep="\n")
print()
print("x_2 (desde el índice 1 hasta el índice 3):", x_2, sep="\n")
print()
print("x_3 (desde el índice 3 hasta el índice 5):", x_3, sep="\n")
print()
print("x_4 (desde el índice 5 hasta el final):", x_4, sep="\n")
````

### Separación de una matriz por columnas

Para realizar la separación se utiliza la función `hsplit` y el corte lo realiza para índices. Esta función recibe como parámetro un matriz y una objeto iterable con los valores de los índices para realizar las separaciones.


````python
x_1, x_2, x_3, x_4 = np.hsplit(matriz_grande, (1, 3, 5))
print("Matriz de trabajo (matriz_grande):", matriz_grande, sep="\n")
print()
print("Sentencia: np.hsplit(matriz_grande, (1, 3, 5))")
print()
print("x_1 (desde el índice 0 hasta el índice 1):", x_1, sep="\n")
print()
print("x_2 (desde el índice 1 hasta el índice 3):", x_2, sep="\n")
print()
print("x_3 (desde el índice 3 hasta el índice 5):", x_3, sep="\n")
print()
print("x_4 (desde el índice 5 hasta el final):", x_4, sep="\n")
````

````python
print("Matriz de trabajo (matriz):", matriz, sep="\n")
matriz_menores_a_4 = np.less(matriz, 4)
matriz_mayores_a_7 = np.greater(matriz, 7)
matriz_modulo_de_2 = np.mod(matriz, 2)
matriz_modulo_de_3 = np.mod(matriz, 3)
matriz_modulo_de_2_igual_0 = np.equal(np.mod(matriz, 2), 0)
matriz_modulo_de_3_igual_0 = np.equal(np.mod(matriz, 3), 0)
print("Matriz de trabajo (matriz_menores_a_4):", matriz_menores_a_4, sep="\n")
print("Matriz de trabajo (matriz_mayores_a_7):", matriz_mayores_a_7, sep="\n")
print("Matriz de trabajo (matriz_modulo_de_2):", matriz_modulo_de_2, sep="\n")
print("Matriz de trabajo (matriz_modulo_de_3):", matriz_modulo_de_3, sep="\n")
print("Matriz de trabajo (matriz_modulo_de_2_igual_0):",
      matriz_modulo_de_2_igual_0, sep="\n")
print("Matriz de trabajo (matriz_modulo_de_3_igual_0):",
      matriz_modulo_de_3_igual_0, sep="\n")
print()
print("Sentencia: np.bitwise_or(matriz_menores_a_4, matriz_mayores_a_7)")
print("Matriz de verdad de valores menores que 4 o mayores a 8:",
      np.bitwise_or(matriz_menores_a_4, matriz_mayores_a_7), sep="\n")
print()
print("Sentencia: np.bitwise_and(matriz_modulo_de_2_igual_0, matriz_modulo_de_3_igual_0)")
print("Matriz de verdad de valores que son múltiplos de 2 y múltiplos 3 ambos a la vez: ",
      np.bitwise_and(matriz_modulo_de_2_igual_0, matriz_modulo_de_3_igual_0), sep="\n")
````

### Función where

La función verifica si se cumple o no con una condición o comparación sobre un numpy array.

Tiene 2 respuestas:
- En caso de pasar un numpy array booleano retorna los indices de los valores que son verdaderos.
- En caso de numpy array y las condiciones `x` e `y` retorna el valor de la posición en `x` si es verdader y el valor de la posición en `y` si es falso.


````python
arreglo_where = np.arange(start=10, stop=22)
print("Arreglo de trabajo (arreglo_where):", arreglo_where, sep="\n")
print()
arreglo_modulo_2 = np.equal(np.mod(arreglo_where, 2), 0)
print("Arreglo de trabajo (arreglo_modulo_2):", arreglo_modulo_2, sep="\n")
print()
print("Sentencia: np.where(arreglo_modulo_2)")
print("Retorna los Índices que cumplen con la condición que sean módulo de 2.",
      np.where(arreglo_modulo_2), sep="\n")
print()
print("Sentencia: np.where(arreglo_menores_a_3, arreglo_where, -1)")
print("Agrego las condiciones x e y para los valores que sean módulo de 2.",
      "- Si la condición se cumple, entonces completa con el valor correspondiente que se está analizando.",
      "- Si la condición NO se cumple, entonces completa con -1",
      np.where(arreglo_modulo_2, arreglo_where, -1), sep="\n")
print()
print("Sentencia: np.where(arreglo_menores_a_3, -1, arreglo_where)")
print("Agrego las condiciones x e y para los valores que sean módulo de 2. ",
      "- Si la condición se cumple, entonces completa con -1",
      "- Si la condición NO se cumple, entonces completa con el valor correspondiente que se está analizando.",
      np.where(arreglo_modulo_2, -1, arreglo_where), sep="\n")
````

## Vectorización de funciones

Estos procesos me permiten vectorizar funciones y por lo tanto las puedo aplicar a los vectores o las matrices de numpy.

````python
# frompyfunc: Vectoriza funciones base de python
base = np.array([2, 3, 4])
exponente = np.array([5, 6, 7])
# array con la función de python pow
# pow recibe como parámetros la base y el exponente
# nin: es la cantidad de parámetros que recibe la función pow
# nout: es la cantidad de resultados que devuelve la función pow
#
vectorizacion_de_la_funcion_pow = np.frompyfunc(pow, nin=2, nout=1)
resultado = vectorizacion_de_la_funcion_pow(base, exponente)

print("Base: \n", base)
print()
print("Exponente: \n", exponente)
print()
print("Resultado: \n", resultado)
print()
print("Operación y resultado:\n")
for b,e,r in zip(base, exponente, resultado):
  print(f"{b} ^ {e} = {r}")
````

````python
# vectorize: Vectoriza cualquier función definida por el usuario
def sqrt(base:float, exponente:float)->float:
  return pow(base, 1/exponente)

base = np.array([25, 216, 2401])
exponente = np.array([2, 3, 4])
# array con la función raiz definida por el usuario
vectorizacion_de_la_funcion_sqrt = np.vectorize(sqrt)
resultado = vectorizacion_de_la_funcion_sqrt(base, exponente)
print("Base: \n", base)
print()
print("Exponente: \n", exponente)
print()
print("Resultado: \n", resultado)
print()
print("Operación y resultado:\n")
for b,e,r in zip(base, exponente, resultado):
  print(f"{b} ^ (1/{e}) = {r}")
````
